```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('IRIS.csv')
df.head()
```

```python
# Pairplot
sns.pairplot(df)
plt.show()
```

```python
# Pairplot
sns.pairplot(df, hue = 'species', palette = 'husl', size = 2)
plt.show()
```

```python
#Create a box plot
sns.boxplot(data = df)

#Add title
plt.title('Distribution of IRIS dataset')
plt.show()
```

```python
# Create a violin plot
sns.violinplot(data=df)

#Add title
plt.title('Violin Plot of IRIS dataset')
plt.show()
```

```python
df = pd.read_csv("IRIS.csv")
df['species'] = df['species'].astype('category').cat.codes
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 7))
sns.scatterplot(x='petal_length', y='petal_width', data=df, hue='species')
plt.show()
```
